```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.distplot(tips['size'])
```

```python
sns.histplot(tips['size'])
```

```python
plt.figure(figsize = (10,8))
sns.set()

sns.distplot(tips['total_bill'], label = 'Total Bill')

plt.title('Histogram of Total Bills')
plt.legend()
```

```python
tips.total_bill.sort_values()
```

```python
bins = [1,5,10,15,20,25,30,35,40,45,50,55]

plt.figure(figsize = (10,8))
sns.set()

sns.distplot(tips['total_bill'], bins = bins)

plt.xticks(bins)

plt.title('Histogram of Total Bills')
# plt.legend()
```
