```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
plt.figure(figsize = (13,10))
sns.scatterplot(x = 'age', y = 'fare', data = titanic, hue = 'sex', style = 'who', size = 'who', sizes = (50,100))
```

```python
plt.figure(figsize = (13,10))
sns.scatterplot(x = 'who', y = 'fare', data = titanic, hue = 'alive', style = 'alive', size = 'who', sizes = (50,100))
```

```python
plt.figure(figsize = (13, 10))

sns.lineplot(x = 'age', y = 'fare', data = titanic)
sns.barplot(x = 'age', y = 'fare', data = titanic)
sns.scatterplot(x = 'age', y = 'fare', data = titanic)
```
