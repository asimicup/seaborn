## Corelation

-Statistical measure of the relationship between two variables

1. The corelation coefficient is the value from -1 to 1
  
-> -1, perfect negative corelation (x-increases when y-decreases)
-> 0, No corelation
-> 1, perfect positive corelation (x-increases when y-increases)

```python
gw_df.corr()
```

```python
plt.figure(figsize = (15, 9))

ax = sns.heatmap(gw_df.corr(), annot = True, linewidth = 3)
ax.tick_params(size = 10, color = 'w', labelsize = 10, labelcolor = 'w')

plt.title('Who is responsible for global warming')
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
cancer_dataset = load_breast_cancer()
cancer_dataset
```

```python
# create dataframe

dataset = pd.DataFrame(np.c_[cancer_dataset['data'], cancer_dataset['target']], 
                      columns = np.append(cancer_dataset['feature_names'], ['target']))

dataset
```

```python
plt.figure(figsize = (25, 25))

heat_map = sns.heatmap(dataset.corr(), annot = True, linewidth = 3)
heat_map.tick_params(size = 10, color = 'w', labelsize = 10, labelcolor = 'w')

plt.title('Breast Cancer')
plt.show()
```
